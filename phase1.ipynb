## Numpy array and basics
numpy

```python
import numpy as np
```

## Creating array from list

```python
arr_1d = np.array([1,2,3,4,5])
print("1D-Array: ",arr_1d)

arr_2d = np.array([[1,2,3], [4,5,6]])
print("2D-Array: ", arr_2d)
```

## List vs Numpy array

```python
py_list = [1,2,3]
print("python list multiplication with 2: ", py_list*2)

np_array = np.array([1,2,3])
print("Numpy array multiplication with 2: ", np_array*2)

# operation time
import time

sl = time.time()
py_list = [i*2 for i in range(1000000)]
print("List operation time: ", time.time()-sl)

sn = time.time()
np_array = np.arange(1000000) * 2
print("Numpy operation time: ", time.time()-sn)
```

## creating array from scratch

```python
zeros = np.zeros((3,4))
print("zeros array: \n", zeros)

ones = np.ones((2,3))
print("Ones array: \n", ones)
```
