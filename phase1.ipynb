## Numpy Array & Basics

```python
import numpy as np
```

#### Creating array from list

```python
arr_1d = np.array([1,2,3,4,5])
print("1D-Array: ",arr_1d)

arr_2d = np.array([[1,2,3], [4,5,6]])
print("2D-Array: ", arr_2d)
```

#### List vs Numpy Array

```python
py_list = [1,2,3]
print("python list multiplication with 2: ", py_list*2)

np_array = np.array([1,2,3])
print("Numpy array multiplication with 2: ", np_array*2)

# operation time
import time

sl = time.time()
py_list = [i*2 for i in range(1000000)]
print("List operation time: ", time.time()-sl)

sn = time.time()
np_array = np.arange(1000000) * 2
print("Numpy operation time: ", time.time()-sn)
```

#### Creating Array from scratch

```python
zeros = np.zeros((3,4))
print("zeros array: \n", zeros)

ones = np.ones((2,3))
print("Ones array: \n", ones)

full = np.full((2,2), 7)
print("Full array: \n", full)

random = np.random.random((2,3))
print("random array: \n", random)

sequence = np.arange(0,10,2)
print("Sequence: \n", sequence)
```

#### Vector, matrix and tensor

```python
vector = np.array([1,2,3])
print("Vector: \n",vector)

matrix = np.array([[1,2,3],
                   [5,6,8]])
print("Matrix: \n", matrix)

tensor = np.array([[1,2],
                  [3,4],
                  [5,6],
                  [7,8]])
print("Tensor: \n", tensor)
```

#### Array Properties

```python
arr = np.array([[1,2,0.3],
                [4,5,6]])

print("Shape: ", arr.shape)
print("Dimension: ", arr.ndim)
print("Size: ", arr.size)
print("DType: ", arr.dtype)
```

#### Array Reshaping

```python
arr = np.arange(12)
print("Original array: \n",arr)

arr_reshaped = arr.reshape((3,4))
print("Reshaped array:\n", arr_reshaped)

flattened = arr_reshaped.flatten()
print("Flattened array: \n", flattened)

raveled = arr_reshaped.ravel()
print("Ravelled array: \n", raveled)

# flatten() creates a copy of the array.
# ravel() returns a view whenever possible (otherwise a copy).

transpose = arr_reshaped.T
print("Transposed array: \n", transpose)
```
