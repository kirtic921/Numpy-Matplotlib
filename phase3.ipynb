## Advance Operations with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure: [restaurent_id, 2021, 2022, 2023, 2024]
sales_data =np.array([
    [1, 150000, 180000, 220000, 250000],  #Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  #Beijing Bites
    [3, 200000, 230000, 260000, 300000],  #Pizza Hub
    [4, 180000, 210000, 240000, 270000],  #Burger Point
    [5, 160000, 185000, 205000, 230000]   #Chai Point
])


print("==== Zomato Data Analysis ====")
```

```python
print("Sales Data Shape: \n", sales_data.shape)
print("Sales data for first three restaurents: \n", sales_data[:3, 1:])
```

```python
# total sales per year
print("Total sales per year: \n", np.sum(sales_data, axis=0))

# in order to remove the id part from beginning
print("Total sales per year excluding the serial no. part: \n", np.sum(sales_data[:, 1:], axis=0))
```

```python
# min sales per restaurent
print("Minimum sales per restaurent: \n", np.min(sales_data[:, 1:], axis=1))

# max sales per year
print("Max sales per year: \n", np.max(sales_data[:, 1:], axis=0))

# avg sales per restaurent
print("Avg sales per restaurent: \n", np.mean(sales_data[:, 1:], axis=1))
```

```python
# cumulative sum
print("Cumulative sum : \n", np.cumsum(sales_data[:, 1:], axis=1))
cumsum = np.cumsum(sales_data[:, 1:], axis=1)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales across all restaurent")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Addition of vectors: \n", vector1+vector2)
print("Multiplication of vectors: \n", vector1*vector2)
print("Dot Product: \n", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("Angle: ", angle)

# np.linalg.norm gives magnitude of vector
```
